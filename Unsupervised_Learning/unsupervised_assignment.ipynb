```python
import pandas as pd
import seaborn as snb 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('customer_data_records.csv')
df.head(10)
```

```python
print(df)
```

```python
dtset = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
by_kmeans = KMeans(n_clusters=2, random_state=0)
by_kmeans.fit(dtset)
```

```python
df['Cluster'] = by_kmeans.labels_
print(df)
```

```python
#visualization
plt.figure(figsize=(15,12))
snb.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster')
plt.title('Customer Expenditure Based on Annual Income')
plt.xlabel('Annual_Income_(k$)')
plt.ylabel('Spending Score(1-100)')
plt.show()
```
