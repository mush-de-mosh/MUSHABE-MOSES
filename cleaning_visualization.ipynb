```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df
```

```python
#empty values or empty cells
#check for empty values
print(df.isnull().sum())
```

```python
#check wrong formats
print(df.dtypes)
```

```python
#wrong data
#check for negative values in numeric columns
for col in ['Duration','Pulse','Maxpulse', 'Calories']:
    if (df[col] < 0).any():
        print(f"Negative values found in column: {col}")
```

```python
#recommendations
#Dont delete empty cells, just fill them
#convert the wrong formats
#correct wrong values
#remove duplicates in rows
```

```python
#fill empty cells with mean for numeric coulumns
for col in ['Duration', 'Pulse', 'Maxpulse', 'Calories']:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
#check shape of our data
print("Shape of the datset:",df.shape)
```

```python
#Summary statistics
print("Statics\n")
df.describe()
```
